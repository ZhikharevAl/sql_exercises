```python
import sqlite3
con = sqlite3.connect('db')
```

```python
import pandas as pd
```

```python
df = pd.DataFrame({'coll':[1, 2, 3]})
```

```python
print(df)
```

```python
df.to_sql('test_table', con, index=False, if_exists='replace')
```

```python
sql = ''' SELECT * FROM test_table t'''
```

```python
pd.read_sql(sql, con)
```

```python
pd.read_sql(sql, con).plot()
```
